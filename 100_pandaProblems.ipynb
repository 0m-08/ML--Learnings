```python
# 26. In the cell below, you have a DataFrame df that consists of 10 columns of floating-point numbers. Exactly 5 entries in each row are NaN values.
# For each row of the DataFrame, find the column which contains the third NaN value.
# You should return a Series of column labels: e, c, d, h, d

nan = np.nan

data = [[0.04,  nan,  nan, 0.25,  nan, 0.43, 0.71, 0.51,  nan,  nan],
        [ nan,  nan,  nan, 0.04, 0.76,  nan,  nan, 0.67, 0.76, 0.16],
        [ nan,  nan, 0.5 ,  nan, 0.31, 0.4 ,  nan,  nan, 0.24, 0.01],
        [0.49,  nan,  nan, 0.62, 0.73, 0.26, 0.85,  nan,  nan,  nan],
        [ nan,  nan, 0.41,  nan, 0.05,  nan, 0.61,  nan, 0.48, 0.68]]

columns = list('abcdefghij')

df26 = pd.DataFrame(data, columns=columns)
(df26.isna().cumsum(axis=1)==3).idxmax(axis=1)
#isna to find nan value, cumsum counts the number of nan, idmax returns the column with max value of nan
```

```python
#27. A DataFrame has a column of groups 'grps' and and column of integer values 'vals':

df27 = pd.DataFrame({'grps': list('aaabbcaabcccbbc'), 
                   'vals': [12,345,3,1,45,14,4,52,54,23,235,21,57,3,87]})
# For each group, find the sum of the three greatest values. You should end up with the answer as follows:

df27= df27.sort_values(['grps','vals'], ascending=[1,0])
df27['count'] = df27.groupby('grps').cumcount() + 1
#groups by grps and cumcount gives running count for each grp starting from 0(default) +1 to start from 1 as highest 2 as mid and 3 as lowest
df27 = df27.pivot_table(columns='grps', values='vals', index='count')
df27

```

```python
# # 28. The DataFrame df constructed below has two integer columns 'A' and 'B'. The values in 'A' are between 1 and 100 (inclusive).
# For each group of 10 consecutive integers in 'A' (i.e. (0, 10], (10, 20], ...), calculate the sum of the corresponding values in column 'B'.
# The answer should be a Series as follows:

# A
# (0, 10]      635
# (10, 20]     360
# (20, 30]     315
# (30, 40]     306
# (40, 50]     750
# (50, 60]     284
# (60, 70]     424
# (70, 80]     526
# (80, 90]     835
# (90, 100]    852
df28=pd.DataFrame(np.random.RandomState(8765).randint(1,101,size=(100,2)),columns=['A','B'])
#randomstate is for getting same random integers everytime , randint generates random int from 1-101 with 100 rows and 2 columns
df28.sort_values(['A'])
groups={i/10:f"({i} , {i+10}]"for i in range (0,100,10)}

df28['A']=(df28["A"]-1)//10
df28["A"]=df28["A"].map(groups)
df28.groupby(['A'])['B'].sum()


```

```python
# 29. Consider a DataFrame df where there is an integer column 'X':

df29 = pd.DataFrame({'X': [7, 2, 0, 3, 4, 2, 5, 0, 3, 4]})
# For each value, count the difference back to the previous zero (or the start of the Series, whichever is closer). These values should therefore be
# [1, 2, 0, 1, 2, 3, 4, 0, 1, 2]
# Make this a new column 'Y'.


# last_zero=np.where(df29['X']==0,df29.index,-1)
# last_zero=pd.Series(last_zero).replace(-1,np.nan).ffill().fillna(-1)
# df29['Y']=df29.index-last_zero
# df29.loc[df29['X']==0,'Y']=0
# df29

x=(df29['X']!=0).cumsum()
y=x!=x.shift()
df29['Y']=y.groupby((y!=y.shift()).cumsum()).cumsum()
df29
```
