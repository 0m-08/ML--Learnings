```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Datasets/data_science_job.csv')
df.head()
```

```python
df.isnull().mean()*100
```

```python
cols=[var for var in df.columns if df[var].isnull().mean()<0.05 and df[var].isnull().mean()>0]
cols
```

```python
new_df=df[cols].dropna()
df.shape, new_df.shape
```

```python
new_df.hist(bins=50, density=True, figsize=(12, 12))
plt.show()
```

```python
import matplotlib.pyplot as plt

# Histogram comparison
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(111)

df['training_hours'].hist(
    bins=50,
    ax=ax,
    density=True,
    color='red',
    alpha=0.5,
    label='Original'
)

new_df['training_hours'].hist(
    bins=50,
    ax=ax,
    density=True,
    color='green',
    alpha=0.5,
    label='After CCA'
)

ax.set_title('Histogram Comparison')
ax.legend()

plt.show()


# Density comparison
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(111)

df['training_hours'].plot.density(
    ax=ax,
    color='red',
    label='Original'
)

new_df['training_hours'].plot.density(
    ax=ax,
    color='green',
    label='After CCA'
)

ax.set_title('Density Plot Comparison')
ax.legend()

plt.show()
```
