# Understanding Your Data 📊

Before diving into Exploratory Data Analysis (EDA) or building Machine Learning models, the very first step is to **understand the structure and characteristics of your dataset**.

In this notebook, we will import and explore our data using standard `pandas` and `numpy` techniques.

### 1. Import Libraries & Load Dataset 📥

We will begin by importing `pandas` and `numpy`, then loading the Titanic `train.csv` dataset located in our `Datasets/` directory.

```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('Datasets/train.csv')
df.head()
```

### 2. How big is the data? 📏

Use `df.shape` to check the number of rows and columns (dimensions of the dataset).

```python
df.shape
```

### 3. How does the data look? 👀

We can inspect random samples of the data using `df.sample()` or look at the last few records using `df.tail()`.

```python
# Get a random sample of 5 rows
df.sample(5)
```

### 4. What are the column data types and memory usage? ℹ️

Use `df.info()` to check data types, non-null counts, and memory usage.

```python
df.info()
```

### 5. Are there any missing values? ❓

Use `df.isnull().sum()` to find the count of missing values in each column.

```python
df.isnull().sum()
```

### 6. How does the data look mathematically? 🧮

Use `df.describe()` to get a statistical summary of the numerical features.

```python
df.describe()
```

### 7. Are there any duplicate rows? 👥

Use `df.duplicated().sum()` to check for duplicate entries.

```python
df.duplicated().sum()
```

### 8. How are the columns correlated? 🔗

Use `df.corr()` on numerical columns to see linear relationships.

```python
# Compute correlation matrix for numeric columns
df.select_dtypes(include='number').corr()
```
